4️⃣ **The output and the reason**
The output of the code is:

[1, 2, 3, 4]

💡 Why?
In Python, lists are mutable, and when you do:

y = x
you are not creating a copy of the list — you are simply creating a new reference to the same list object in memory.

So when you do:

y.append(4)
you’re modifying the list that both x and y refer to. Hence, printing x also shows the updated list.

✅ To avoid this (create a copy):
You could use:

y = x.copy()

or

y = list(x)

or
y = x[:]
Then x and y would refer to two different list objects.
